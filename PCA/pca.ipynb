```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
df = make_blobs(n_samples=1000, n_features=3, cluster_std=5)
df = pd.DataFrame(df[0], columns=['a','b','c'])
```

```python
%matplotlib inline
ax = plt.axes(projection='3d')
ax.scatter(df['a'], df['b'], df['c'])
```

```python
model = PCA(n_components=2)
pcadf = model.fit_transform(df)
plt.scatter(pcadf[:,0], pcadf[:,1])
```
